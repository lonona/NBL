#### Load data using pandas

### Visualise a kernel density estimate (kde) of the data

The univariate distribution of the observation clearly shows that the data contains multiple clusters

### find the best mixture of Gaussians for the data

```python
AIC, BIC, models = gmm_func(data=df['NO3(mg/l)'].values, ncomponents=5, logscale=True)
```

* generate the best-fit mixture of the data

```python
X = df['NO3(mg/l)'].values.reshape(-1, 1)
X = np.log1p(X.copy())

plt.figure(figsize=(12, 10), dpi = 100)
M_best = models[np.argmin(AIC)]

x = np.linspace(X.min() - 1, X.max() + 1, 403)
logprob = M_best.score_samples(x.reshape(-1, 1))
responsibilities = M_best.predict_proba(x.reshape(-1, 1))
pdf = np.exp(logprob)
pdf_individual = responsibilities * pdf[:, np.newaxis]

plt.hist(X, 30, density=True, histtype='stepfilled', alpha=0.4)
plt.plot(x, pdf, '-k')
plt.plot(x, pdf_individual, '--k')
plt.title("Best-fit Mixture")
plt.xlabel(r'$\log[x]$')
plt.ylabel('$p(x)$')
plt.ylim(0, 0.8)
```
